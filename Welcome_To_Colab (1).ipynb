```
def add_expense():
    category = input("Enter category: ")
    amount = input("Enter amount: ")

    file = open("expenses.csv", "a")
    file.write(category + "," + amount + "\n")
    file.close()

    print("Expense added successfully!")


def view_expense():
    try:
        file = open("expenses.csv", "r")
        print("\nYour Expenses:")

        for line in file:
            print(line.strip())

        file.close()
    except:
        print("No expenses found!")


def total_expense():
    try:
        file = open("expenses.csv", "r")
        total = 0

        for line in file:
            data = line.split(",")
            total += int(data[1])

        file.close()
        print("Total Expense =", total)
    except:
        print("No data available!")


def menu():
    while True:
        print("\n--- Expense Tracker ---")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Total Expense")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expense()
        elif choice == "3":
            total_expense()
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice")


menu()
```
